```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv("honeyproduction.csv")
```

### Analysing the data

```python
print(df.head())

prod_per_year= df.groupby('year').totalprod.mean().reset_index()

X = prod_per_year["year"]
X= X.values.reshape(-1, 1)
#print(X)
```

```python
Y = prod_per_year["totalprod"]

plt.scatter(X, Y)
plt.show()
```

### Creating and fitting a Linear Regression model

```python
regr= linear_model.LinearRegression()
regr.fit(X, Y)
print("slope of line = " + str(regr.coef_[0]))
print("intecept of line = " + str(regr.intercept_))

y_predict= regr.predict(X)

plt.plot(X, y_predict)
plt.scatter(X, Y)
plt.show()
```

### Predicting Honey Decline:
##### Predicting decline for the year 2050

```python
X_future= np.array(range(2013, 2051))
X_future= X_future.reshape(-1, 1)

#var below predicts future y val for X_future 
future_predict = regr.predict(X_future)
plt.scatter(X, Y)
plt.plot(X, y_predict)
plt.plot(X_future, future_predict)
plt.show()

print("Predicted total production of honey (in pounds) for the year 2050: " + str(future_predict[-1]))
```
